### 1.Maximum product of two elements in an array

```python
nums = [3, 4, 5, 2]
nums.sort()
print(nums)
result = (nums[-1] - 1) * (nums[-2] - 1)
print(result)
```

### 2.Count number of Teams

```python
rating = [2, 5, 3, 4, 1]
n = len(rating)
up = [0] * n
down = [0] * n
teams = 0
for i in range(n - 1, -1, -1):
    for j in range(i + 1, n):
        if rating[i] < rating[j]:
            up[i] += 1
            teams += up[j]
        else:
            down[i] += 1
            teams += down[j]
print(teams)
```

### 3.Number of Students Doing Homework at a Given Time

```python
start_time = [1, 2, 3]
end_time = [3, 2, 7]
query_time = 4
count = 0
for i in range(len(start_time)):
    if start_time[i] <= query_time and query_time <= end_time[i]:
        count = count + 1
print(count)
```

### 4.Number of steps to Reduce a number to zeroes

```python
num = 14
steps = 0

while num != 0:
    steps = steps + 1
    if num % 2 == 1:
        num = num - 1
    else:
        num = num // 2
print(steps)
```

### 5.Counting Bits

```python
n = 5
result = []
for i in range(n + 1):
    ones_count = bin(i).count("1")
    result.append(ones_count)
print(result)
```
