### 1.Defanging an IP Address

```python
a = "1.1.1.1"
ans = ""
for i in a:
    if i != "." :
        ans = ans + i
    else:
        ans = ans + "[.]"
print(ans)
```

### 4. How many numbers are smaller than current number

```python
nums = [8, 1, 2, 2, 3]
n = len(nums)
answer = []
for i in range(n):
    count = 0
    for j in range(n):
        if nums[i] > nums[j]:
            count = count + 1
    answer.append(count)
print(answer)
```

### 5. Subtract the Product and Sum of Digits of an Integer

```python
n = 234
sum_digits = 0
pdt_digits = 1
for ch in str(n):
    digit = int(ch)
    sum_digits = sum_digits + digit
    pdt_digits = pdt_digits * digit
result = pdt_digits - sum_digits
print(result)
```

### 6. XOR Operation in an Array
